```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("../data/processed/processed_heart_disease_v0.csv", sep=",")
df.head(5)
```

```python
print("\nStatistiques descriptives:")
df.describe()
```

```python
# Distribution de la variable cible

target_col = 'Heart Disease'
print(df[target_col].value_counts())
print(f"\nProportion: {df[target_col].value_counts(normalize=True)}")
```

```python
# Visualisation
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
df[target_col].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Distribution de la Variable Cible')
plt.xlabel('Classe (0=Sain, 1=Malade)')
plt.ylabel('Nombre')
```

```python
plt.subplot(1, 2, 2)
df[target_col].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'salmon'])
plt.title('Proportion des Classes')
```

```python
plt.figure(figsize=(14, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Matrice de Corrélation')
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.show()
```

```python
print("\nCorrélations avec la cible:")
print(correlation_matrix[target_col].sort_values(ascending=False))
```
